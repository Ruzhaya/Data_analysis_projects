# Исследование рынка компьютерных игр

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

# Описание данных

* **Name** — название игры
* **Platform** — платформа
* **Year_of_Release** — год выпуска
* **Genre** — жанр игры
* **NA_sales** — продажи в Северной Америке (миллионы проданных копий)
* **EU_sales** — продажи в Европе (миллионы проданных копий)
* **JP_sales** — продажи в Японии (миллионы проданных копий)
* **Other_sales** — продажи в других странах (миллионы проданных копий)
* **Critic_Score** — оценка критиков (максимум 100)
* **User_Score** — оценка пользователей (максимум 10)
* **Rating** — рейтинг от организации *ESRB* (англ. *Entertainment Software Rating Board*). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Заменяю значение `tbd` на `nan`, т.к. аббревиатура **TBD** расшифровывается как *To Be Determined (будет определено) / To Be Decicded (будет решено)*, т.е. на момент выгрузки данных пользователи ещё не оценили указанные игры. <br>

Заменить пропуски в данном столбце не представляется возможным, т.к. оценка пользователя субъективна и в большей степени зависит от платформы, на которой вышла игра.

```python
df['user_score'] = (df['user_score']
                    .replace('tbd', np.nan)
                    .astype('float64')
                    )
df['user_score'].unique()
```

### Обработка столбца `rating`

```python
df['rating'].unique()
```

Так как **Entertainment Software Rating Board** *(ESRB)* в **1998** году рейтинг `K-A` *(Kids to Adults)* поменяли на `E` *(Everyone)* меняю значения для устранения неявных дублей. 

Пропуски в рейтинге для одноименных игр заменяю на первое значение рейтинга, остальные заменяю на `Unknow`.

```python
df['rating'] = (df['rating']
                .fillna(df.groupby('name')['rating'].transform('first'))
                .replace('K-A', 'E')
               ).fillna('Unknow')
df['rating'].unique()
```

## Проверка гипотез

### Первая гипотеза

Средние пользовательские рейтинги платформ *Xbox One* и *PC* одинаковые

#### Формулируем нулевую $H_0$ и альтернативную $H_1$ гипотезы

$H_0$ - средние пользовательские рейтинги платформ Xbox One и PC равны <br>
$H_1$ - средние пользовательские рейтинги платформ не равны

#### Формируем выборки для проверки

```python
df_xone = df_actual['user_score'].loc[df_actual['platform'] == 'XOne']

df_pc = df_actual['user_score'].loc[df_actual['platform'] == 'PC']
```

#### Функция расчета дисперсии и построения распределения

```python
def var_distr(df1, name1, df2, name2):
    print('Дисперсия выборки ' + name1 + ':', round(np.var(df1, ddof=1), 2), end='\n')
    display(df1.describe())
    sns.displot(data=df1, bins=20)
    plt.title('График распределения оценок ' + name1)
    
    print('Дисперсия выборки ' + name2 + ':', round(np.var(df2, ddof=1), 2), end='\n')
    display(df2.describe())
    sns.displot(data=df2, bins=20)
    plt.title('График распределения оценок ' + name2)
    plt.show()
```

#### Проверяем дисперсию и распределения выборок

```python
var_distr(df_xone, 'XOne', df_pc, 'PC')
```

#### Функция для проверки гипотез

```python
def hyp(df1, df2, alp, equal_var):
    alpha = alp
    result = st.ttest_ind(df1, df2, equal_var = equal_var, nan_policy='omit')

    print('p-значение:', result.pvalue)

    if result.pvalue < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не получилось отвергнуть нулевую гипотезу')
```

#### Проверяем гипотезу

```python
hyp(df_xone, df_pc, .01, True)
```

#### Вывод

В результате проверки **не получилось отвергнуть** нулевую гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC равны. 

### Вторая гипотеза

Средние пользовательские рейтинги жанров *Action* (англ. «действие», экшен-игры) и *Sports* (англ. «спортивные соревнования») разные.

#### Формулируем нулевую $H_0$ и альтернативную $H_1$ гипотезы

$H_0$ - средние пользовательские рейтинги жанров Action и Sports одинаковые <br>
$H_1$ - средние пользовательские рейтинги указанных жанров разные

#### Формируем выборки для проверки

```python
df_action = df_actual['user_score'].loc[df_actual['genre'] == 'Action']

df_sports = df_actual['user_score'].loc[df_actual['genre'] == 'Sports']
```

#### Проверяем дисперсию и распределения выборок

```python
var_distr(df_action, 'Action', df_sports, 'Sports')
```

#### Проверяем гипотезу

```python
hyp(df_action, df_sports, 0.01, False)
```

#### Вывод

Проведя T-test и получив вероятность $1,4e^-20$ мы **вынуждены отвергнуть** нулевую гипотезу о том, что средние пользовательские рейтинги жанров Action и Sports равны.
