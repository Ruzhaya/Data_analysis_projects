`game_actions.csv`

    * `spaceport` - космопорт

* `project_type` - тип реализованного проекта

`ad_cost.csv`
* `day` - день, в который был совершён клик по объявлению
* `source` - источник трафика
* `cost` - стоимость кликов

`user_source.csv`
* `user_id` - идентификатор пользователя
* `source` - источник, с которого пришёл пользователь, установивший приложение

```python
import pandas as pd
import numpy as np
import math
from datetime import datetime, timedelta
import seaborn as sns
from matplotlib import pyplot as plt
from scipy import stats as st
```

### Функции

```python
def profiles_orders(sessions, sources, ad):
    """Formation of user profiles"""
    cogort = (
        sessions.sort_values(by=['user_id', 'event_datetime'])
        .groupby('user_id')
        .agg(
            {
                'event_datetime':'first'    
            }
        )
        .rename(columns={'event_datetime':'first_ts'})
        .sort_values(by='first_ts')
        .reset_index()
    )
        
    last_event = (
        sessions.sort_values(by=['user_id', 'event_datetime'])
        .groupby('user_id')
        .agg(
            {
                'event_datetime':'last',
                'event':'last'
            }
        )
        .rename(columns={'event_datetime':'last_ts', 'event':'last_event'})
        .sort_values(by='last_ts')
        .reset_index()
    )
    
    last_build = (
        sessions.query('building_type != "None"')
        .sort_values(by=['user_id', 'event_datetime'])
        .groupby('user_id')
        .agg(
            {
                'building_type':'last',
                'event_datetime':'last'
            }
        )
        .rename(columns={'building_type':'last_build', 'event_datetime':'last_build_dt'})
        .sort_values('last_build_dt')
        .reset_index()
    )
    
        
        
    profiles = (cogort
                .merge(last_event, 
                       on='user_id', 
                       how='left')
                .merge(source, 
                       on='user_id', 
                       how='left')
                .merge(last_build,
                      on='user_id',
                      how='left')
               )
    profiles['dt'] = profiles['first_ts'].dt.date
    profiles['last_ts'] = profiles['last_ts'].dt.date
    profiles['last_build_dt'] = profiles['last_build_dt'].dt.date
    profiles['lifetime_user'] = (profiles['last_ts'] - profiles['dt']).dt.days
    profiles['building'] = (profiles['last_build_dt'] - profiles['dt']).dt.days
        
    ad['users_cogort'] = ad['day'] + timedelta(1)
    ad = ad.drop(columns='day')
    
    df = profiles[['user_id', 'dt']]
    df = df.merge(source, on='user_id', how='left')
    df = (df
          .pivot_table(index=['source', 'dt'], 
                       values='user_id', 
                       aggfunc='count')
          .reset_index()
          .rename(columns={'source':'sr', 
                           'user_id':'cogort_value'})
         )
    df = pd.concat([df, ad], axis=1).drop(columns=['source', 'dt'])
    df['users_cogort'] = df['users_cogort'].dt.date
    df['cost_per_user'] = round((df['cost']/df['cogort_value']), 3)
    
    profiles = profiles.merge(df, left_on=['source', 'dt'], right_on=['sr', 'users_cogort'])
    profiles = profiles.drop(columns=['sr','users_cogort'], axis=1)
    
    orders = (sessions
              .query('event == "building"')[['event_datetime', 'event', 'user_id', 'building_type']]
              .pivot_table(index=['user_id','event_datetime', 'building_type'], values='event', aggfunc='count')
              .rename(columns={'event':'event_count'})
              .reset_index()
             )
    orders['revenue'] = orders['event_count']*0.07
    orders = orders.drop(columns='event_count')
   
    return profiles, orders
```

```python
def get_ltv(
    profiles,
    purchases,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):
    """Calculation of LTV & ROMI metrics"""

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    # добавляем данные о покупках в профили
    result_raw = result_raw.merge(
        purchases, on='user_id', how='left'
    )
    # рассчитываем лайфтайм пользователя для каждой покупки
    result_raw['lifetime'] = (
        result_raw['event_datetime'] - result_raw['first_ts']
    ).dt.days
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']
        

    # функция группировки по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        # строим «треугольную» таблицу выручки
        result = df.pivot_table(
            index=dims, columns='lifetime', values='revenue', aggfunc='sum'
        )
        # находим сумму выручки с накоплением
        result = result.fillna(0).cumsum(axis=1)
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        
        # объединяем размеры когорт и таблицу выручки
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # считаем LTV: делим каждую «ячейку» в строке на размер когорты
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes

        # собираем датафрейм с данными пользователей и значениями CAC, 
        # добавляя параметры из dimensions
        cac = df[['user_id', 'cost_per_user'] + dims].drop_duplicates()

        # считаем средний CAC по параметрам из dimensions
        cac = (
            cac.groupby(dims)
            .agg({'cost_per_user': 'mean'})
            .rename(columns={'cost_per_user': 'cac'})
        )

        # считаем ROI: делим LTV на CAC
        romi = result.div(cac['cac'], axis=0)

        # удаляем строки с бесконечным ROI
        romi = romi[~romi['cohort_size'].isin([np.inf])]

        # восстанавливаем размеры когорт в таблице ROI
        romi['cohort_size'] = cohort_sizes

        # добавляем CAC в таблицу ROI
        romi['cac'] = cac['cac']

        # в финальной таблице оставляем размеры когорт, CAC
        # и ROI в лайфтаймы, не превышающие горизонт анализа
        romi = romi[['cohort_size', 'cac'] + list(range(horizon_days))]

        # возвращаем таблицы LTV и ROI
        return result, romi

    # получаем таблицы LTV и ROI
    result_grouped, romi_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )

    # для таблиц динамики убираем 'cohort' из dimensions
    if 'cohort' in dimensions:
        dimensions = []

    # получаем таблицы динамики LTV и ROI
    result_in_time, romi_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    return (
        result_raw,  # сырые данные
        result_grouped,  # таблица LTV
        result_in_time,  # таблица динамики LTV
        romi_grouped,  # таблица ROI
        romi_in_time,  # таблица динамики ROI
    ) 
```

```python
def filter_data(df, window):
    """Frame smoothing function"""
    # для каждого столбца применяем скользящее среднее
    for column in df.columns.values:
        df[column] = df[column].rolling(window).mean() 
    return df 
```

```python
def plot_ltv_romi(ltv, ltv_history, romi, romi_history, horizon, window=1):
    """Function to visualize LTV and ROI"""

    # задаём сетку отрисовки графиков
    plt.figure(figsize=(20, 15))

    # из таблицы ltv исключаем размеры когорт
    ltv = ltv.drop(columns=['cohort_size'])
    # в таблице динамики ltv оставляем только нужный лайфтайм
    ltv_history = ltv_history.drop(columns=['cohort_size'])[[horizon - 1]]

    # стоимость привлечения запишем в отдельный фрейм
    cac_history = romi_history[['cac']]

    # из таблицы roi исключаем размеры когорт и cac
    romi = romi.drop(columns=['cohort_size', 'cac'])
    # в таблице динамики romi оставляем только нужный лайфтайм
    romi_history = romi_history.drop(columns=['cohort_size', 'cac'])[
        [horizon - 1]
    ]

    # первый график — кривые ltv
    ax1 = plt.subplot(2, 3, 1)
    ltv.T.plot(grid=True, ax=ax1)
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('LTV')

    # второй график — динамика ltv
    ax2 = plt.subplot(2, 3, 2, sharey=ax1)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in ltv_history.index.names if name not in ['dt']]
    filtered_data = ltv_history.pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax2)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика LTV пользователей на {}-й день'.format(horizon))

    # третий график — динамика cac
    ax3 = plt.subplot(2, 3, 3, sharey=ax1)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in cac_history.index.names if name not in ['dt']]
    filtered_data = cac_history.pivot_table(
        index='dt', columns=columns, values='cac', aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.title('Динамика стоимости привлечения пользователей')

    # четвёртый график — кривые romi
    ax4 = plt.subplot(2, 3, 4)
    romi.T.plot(grid=True, ax=ax4)
    plt.axhline(y=1, color='red', linestyle='--', label='Уровень окупаемости')
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('ROMI')

    # пятый график — динамика romi
    ax5 = plt.subplot(2, 3, 5, sharey=ax4)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in romi_history.index.names if name not in ['dt']]
    filtered_data = romi_history.pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax5)
    plt.axhline(y=1, color='red', linestyle='--', label='Уровень окупаемости')
    plt.xlabel('Дата привлечения')
    plt.title('Динамика ROMI пользователей на {}-й день'.format(horizon))

    plt.tight_layout()
    plt.show()
```

```python
def var_distr(df1, name1, df2, name2, bins):
    
    """Variance calculation and distribution check"""
    
    print('Дисперсия выборки ' + name1 + ':', round(np.var(df1, ddof=1), 2), end='\n')
    display(df1.describe())
    print('-'*100)
    results_1 = st.shapiro(df1)
    p_value_1 = results_1[1]
    alpha = 0.05
    print('\033[1m Критерий Шапиро-Уилка \033[0m')
    print('p-значение: ', p_value_1)
    if p_value_1 > alpha:
        print('Не получилось отвергнуть нулевую гипотезу, всё нормально')
    else:
        print('Отвергаем нулевую гипотезу: распределение не нормально')
    
    print('-'*100)
    sns.histplot(x=df1, bins=bins)
    plt.title('График распределения выборки ' + name1)
    plt.show()
    print('*'*100)
    print('Дисперсия выборки ' + name2 + ':', round(np.var(df2, ddof=1), 2), end='\n')
    display(df2.describe())
    print('-'*100)
    results_2 = st.shapiro(df2)
    p_value_2 = results_2[1]
    alpha = 0.05
    print('\033[1m Критерий Шапиро-Уилка \033[0m')
    print('p-значение: ', p_value_2)
    if p_value_2 > alpha:
        print('Не получилось отвергнуть нулевую гипотезу, всё нормально')
    else:
        print('Отвергаем нулевую гипотезу: распределение не нормально')
    print('-'*100)
    sns.histplot(x=df2, bins=bins)
    plt.title('График распределения выборки ' + name2)
    plt.show()
    print('*'*100)
```

```python
def hyp_mannwhitney(df1, df2, alp, equal_var):
    
    """Mann-Whitney check"""
    
    alpha = alp

    results = st.mannwhitneyu(df1, df2, alternative='two-sided')

    print('p-значение: ', results.pvalue)

    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу: разница статистически значима')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

```python
def z_test(cogort, project, battle, alpha):
    
    """Z-test check"""
    
    for i in range(len(cogort)):
        p_pro = project[i] / cogort[i]
        p_battle = battle[i] / cogort[i]
        p_combined = (project[i] + battle[i]) / (2*cogort[i])
        difference = p_pro - p_battle
        z_value = difference / math.sqrt(p_combined * (1 - p_combined) * (2/cogort[i]))
        distr = st.norm(0, 1)
        p_value = (1 - distr.cdf(abs(z_value))) * 2
        
        print('Размер когорты:\033[1m {}\033[0m'.format(cogort[i]))
        print('Количество пользователей группы "проект":\033[1m {}\033[0m'.format(project[i]))
        print('Количество пользователей группы "победа":\033[1m {}\033[0m'.format(battle[i]))
        print('')
        print('p-значение: ', p_value)

        if p_value < alpha:
            print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
        else:
            print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными')
        print('-'*100)
```

## Проверка гипотез

### Первая гипотеза

Гипотеза о различии времени прохождения уровня между пользователями, которые заканчивают уровень через реализацию проекта, и пользователями, которые заканчивают уровень победой над другим игроком.

#### Формулирование нулевой и альтернативной гипотез

$H_0$ - время прохождения уровня для пользоваетелей, заканчивающих уровень через реализацию проекта и для пользователей, заканчивающих уровень победой над другим игроком, равно <br>
$H_1$ - время прохождения уровня различается между стратегиями

#### Рассчет дисперсии и распределения исследуемых выборок

```python
var_distr(project_profiles['lifetime_user'], 'Проект', battle_profiles['lifetime_user'], 'Победа', 25)
```

#### Проверка гипотезы

```python
hyp_mannwhitney(project_profiles['lifetime_user'], battle_profiles['lifetime_user'], .01, True)
```

#### Вывод

Так как проверка критерием **Шапиро-Уилка** показала, что распределение выборок не нормальное, было принято решение проверять гипотезу непараметрическим тестом **Уилкоксона-Манна-Уитни**. По результатам проверки мы вынуждены **отвергнуть нулевую гипотезу** о равенстве времени прохождения уровня

### Вторая гипотеза

Гипотеза о различии долей между количеством пользователей, реализующих проект и количеством пользователей, побеждающих противника от общего количества пользователей в каждой когорте.

#### Формулирование нулевой и альтернативной гипотез

$H_0$ - доля пользователей, завершающих уровень посредством реализации проекта от общего количества пользователей когорты равна доле пользователей, завершающих уровень посредством победы над другим игроком <br>
$H_1$ - доли пользователей не равны

#### Формирование выборок

##### Формирование списка размеров когорт

```python
cohort = (actions
          .groupby('user_id')
          .agg({'event_date':'first'})
          .reset_index()
          .pivot_table(index='event_date', 
                       values='user_id', 
                       aggfunc='count')
          .reset_index()['user_id']
         )
cohort
```

##### Выборка количества пользователей, завершивших уровень через проект

```python
project_1 = (
    project_profiles
    .pivot_table(index='dt',
                values='user_id',
                aggfunc='count')
    .reset_index()['user_id']
)
project_1
```

##### Выборка количества пользователей, завершивших уровень победой

```python
battle_1 = (
    battle_profiles
    .pivot_table(index='dt',
                values='user_id',
                aggfunc='count')
    .reset_index()['user_id']
)
battle_1
```

#### Проверка гипотезы

```python
z_test(cohort, project_1, battle_1, 0.01)
```

#### Вывод

Для проверки гипотезы о равенстве долей был выбран **z-тест**, сформировав выборки с количеством пользователей по когортам было проведено 7 тестов, показавших, что между долями выборок в каждой когорте есть значимая разница, нулевая гипотеза отвергается.

### Третья гипотеза

Гипотеза о равенстве среднего количества зданий для группы игроков, завершивших уровень проектом и для группы игроков, завершивших уровень победой.

#### Формулирование нулевой и альтернативной гипотез

$H_0$ - среднее количество зданий у пользователей, завершающих уровень посредством реализации проекта равно среднему количеству зданий у пользователей, завершающих уровень посредством победы над другим игроком <br>
$H_1$ - средние значения групп не равны

#### Проверка выборок на нормальность распределения

```python
var_distr(project_profiles['building'], 'Проект', battle_profiles['building'], 'Победа', 25)
```

#### Проверка гипотезы

```python
hyp_mannwhitney(project_profiles['building'], battle_profiles['building'], .01, True)
```

#### Вывод

Проверив выборки для третьей гипотезы на равномерность так же как и в первом случае было принято решение использовать тест **Уилкоксона-Манна-Уитни**, который показал, что разница между среднем количеством строений группы "проект" и средним количеством строений группы "победа" статистически значима, нулевая гипотеза отвегается.
